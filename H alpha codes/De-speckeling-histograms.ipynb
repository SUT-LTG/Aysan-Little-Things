```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file
df = pd.read_csv(r"C:\Users\AYSAN\Desktop\project\Galaxy\H-alpha regions\different methods results.csv")
lee_4 = df['lee_4'].tolist()
kuan_4 = df['kuan_4'].tolist()
nl_means_4 = df['nl_means_4'].tolist()

# Plot the histograms with different colors and transparency
plt.hist(lee_4, bins=30, color='blue', alpha=0.3, label='lee_4 Histogram', edgecolor='black')
plt.hist(kuan_4, bins=30, color='red', alpha=0.3, label='kuan_4 Histogram', edgecolor='black')
plt.hist(nl_means_4, bins=30, color='green', alpha=0.3, label='nl_means_4 Histogram', edgecolor='black')
# Add title and legend
plt.title("Level 4")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# Plot the smooth KDE plots with different colors, transparency, and adjusted bandwidth
sns.kdeplot(lee_4, bw_adjust=0.05, color='blue', fill=False, alpha=0.5, label='lee_4 KDE')
sns.kdeplot(kuan_4, bw_adjust=0.05, color='red', fill=False, alpha=0.5, label='kuan_4 KDE')
sns.kdeplot(nl_means_4, bw_adjust=0.05, color='green', fill=False, alpha=0.5, label='nl_means_4 KDE')

# Add title and legend
plt.title("Level 4")
plt.xlabel("Value")
plt.ylabel("Probability Density Function (PDF)")
plt.legend()
plt.show()
```
