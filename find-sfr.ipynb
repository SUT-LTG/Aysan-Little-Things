```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define the file path to your CSV file.
file_path = r"C:\Users\AYSAN\Desktop\project\Galaxy\H-alpha regions\Corrected_H-alpha_analysis + paper values.csv"

# Read the CSV file into a DataFrame.
df = pd.read_csv(file_path)

# Compute L-H-alpha from "paper logH-alpha" with correct unit adjustment.
# The conversion is applied after the exponentiation.
if "L-H-alpha" not in df.columns:
    df["L-H-alpha"] = 10 ** df["paper logH-alpha"] * 1e-7

# Convert Intrinsic Flux to Intrinsic Luminosity.
# 1 Mpc = 3.086e22 meters.
# Luminosity is computed as: L = F * 4 * pi * d^2, where d is in meters.
df["Intrinsic Luminosity"] = df["Intrinsic Flux"] * 4 * np.pi * (df["D"] * 3.086e22)**2

# Save the updated DataFrame back to the CSV file, so the new column is stored.
df.to_csv(file_path, index=False)

# Define variables:
# L_paper: The luminosity derived from paper logH-alpha ("Paper Luminosity")
# L_intrinsic: The intrinsic luminosity computed from intrinsic flux and distance.
L_paper = df["L-H-alpha"].values
L_intrinsic = df["Intrinsic Luminosity"].values

# Compute the best-fit coefficient 'a' for the model:
# Paper Luminosity = a * Intrinsic Luminosity
a = np.sum(L_paper * L_intrinsic) / np.sum(L_intrinsic**2)
print("Best-fit coefficient a (where Paper Luminosity = a * Intrinsic Luminosity) is:", a)

# For the fit line, rearrange the relation:
# Intrinsic Luminosity = Paper Luminosity / a
x_fit = np.linspace(min(L_paper), max(L_paper), 100)
y_fit = x_fit / a

# Create the scatter plot.
plt.figure(figsize=(10, 8))
plt.scatter(L_paper, L_intrinsic, color="blue", alpha=0.7, label="Data Points")

# Annotate each data point with its "Number" (ensure the column name is correct).
for _, row in df.iterrows():
    plt.annotate(
        str(row["Number"]),
        (row["L-H-alpha"], row["Intrinsic Luminosity"]),
        textcoords="offset points",
        xytext=(5, 5),
        ha="center"
    )

# Plot the best-fit line.
plt.plot(x_fit, y_fit, color="red", lw=2, label=f"Best-Fit: Intrinsic Luminosity = Paper L-H-alpha / a, a = {a:.2e}")

# Label the axes and add a title.
plt.xlabel("Paper L-H-alpha (Luminosity in Watts)")
plt.ylabel("Intrinsic Luminosity (ADU * m^2)")
plt.title("Scatter Plot of Intrinsic Luminosity vs Paper Luminosity")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define the file path to your CSV file.
file_path = r"C:\Users\AYSAN\Desktop\project\Galaxy\H-alpha regions\Corrected_H-alpha_analysis + paper values.csv"

# Read the CSV file into a DataFrame.
df = pd.read_csv(file_path)

# Compute L-H-alpha from "paper logH-alpha" with correct unit adjustment.
if "L-H-alpha" not in df.columns:
    df["L-H-alpha"] = 10 ** df["paper logH-alpha"] * 1e-7

# Drop rows 17, 18, 19, 20, and 21 (indices start at 0).
rows_to_remove = [17, 18, 19, 20, 21]
df_reduced = df.drop(index=rows_to_remove)

# Define variables for the reduced dataset.
L = df_reduced["L-H-alpha"].values
F = df_reduced["Intrinsic Luminosity"].values

# Compute the best-fit coefficient 'a' for the reduced dataset:
a = np.sum(L * F) / np.sum(F**2)
print("Best-fit coefficient a (Excluding Rows 17-21):", a)

# For the plot, the best-fit line:
# Intrinsic Luminosity = L-H-alpha / a
x_fit = np.linspace(min(L), max(L), 100)
y_fit = x_fit / a

# Create scatter plot (excluding rows 17-21).
plt.figure(figsize=(10, 8))
plt.scatter(L, F, color="blue", alpha=0.7, label="Data Points (Excluding Rows 17-21)")

# Annotate each data point with its "Number" column (ensure the column name is correct).
for _, row in df_reduced.iterrows():
    plt.annotate(
        str(row["Number"]),  # Adjust annotation placement
        (row["L-H-alpha"], row["Intrinsic Luminosity"]),
        textcoords="offset points",
        xytext=(10, -10),  # Offset the text for better readability
        ha="center",
        fontsize=8,  # Reduce font size if needed for clarity
        color="darkred"  # Change text color to differentiate it
    )

# Plot the best-fit line.
plt.plot(x_fit, y_fit, color="red", lw=2, label=f"Best-Fit Line: Luminosity = L-H-alpha / {a:.2e}")

# Label the axes and add a title.
plt.xlabel("L-H-alpha (Paper Luminosity in Watts)")  # Clear specification of the x-axis variable
plt.ylabel("Intrinsic Luminosity (ADU * m^2)")  # Clear specification of the y-axis variable
plt.title("Scatter Plot of Intrinsic Luminosity vs L-H-alpha from the paper (Excluding Rows 17-21)")
plt.legend()
plt.grid(True)
plt.show()

a_for_some_data =  a
```
